```python
class myDataset(Dataset):

	def __init__(self,
			   df               = None,
			   scaler_area      = None,
			   scaler_bedrooms  = None,
			   scaler_bathrooms = None,
			   scaler_stories   = None,
			   scaler_price     = None, 
			   **kwargs
			   ):

		self.df                     = df
		self.df["scaled area"]      = scaler_area.transform(np.asarray(df["area"]).reshape(-1,1))
		self.df["scaled bedrooms"]  = scaler_bedrooms.transform(np.asarray(df["bedrooms"]).reshape(-1,1))
		self.df["scaled bathrooms"] = scaler_bathrooms.transform(np.asarray(df["bathrooms"]).reshape(-1,1))
		self.df["scaled stories"]   = scaler_stories.transform(np.asarray(df["stories"]).reshape(-1,1))
		self.df["scaled price"]     = scaler_price.transform(np.asarray(df["price"]).reshape(-1,1))

	def __len__(self):
		return len(self.df)
	
	def __getitem__(self,idx = None, **kwargs):
		input = torch.tensor(self.df.iloc[idx,5:-1].values)
		label = torch.tensor(self.df.iloc[idx,-1])
		return input,label
		
```

```python
scaler_area.fit(np.asarray(df_data["area"]).reshape(-1,1))
scaler_bedrooms.fit(np.asarray(df_data["bedrooms"]).reshape(-1,1))
scaler_bathrooms.fit(np.asarray(df_data["bathrooms"]).reshape(-1,1))
scaler_stories.fit(np.asarray(df_data["stories"]).reshape(-1,1))
scaler_price.fit(np.asarray(df_data["price"]).reshape(-1,1))
```

```python
train_dataset = myDataset(df_train,scaler_area,scaler_bedrooms,scaler_bathrooms,scaler_stories,scaler_price)
test_dataset = myDataset(df_test,scaler_area,scaler_bedrooms,scaler_bathrooms,scaler_stories,scaler_price)
```

```python
test_dataset[0]
```

```python
train_loader = DataLoader(train_dataset,batch_size=64)
test_loader  = DataLoader(test_dataset,batch_size=64)
```

```python
class UANet(torch.nn.Module):
	def __init__(self , input_size = None, hidden_size = None,**kwargs):
		super(UANet,self).__init__()

		self.relu = torch.nn.ReLU()

		self.fc1 = torch.nn.Linear(input_size,hidden_size)
		self.fc2 = torch.nn.Linear(hidden_size,1)

	def forward(self, x):
		output = self.fc1(x)
		output = self.relu(output)
		output = self.fc2(output)

		return output
		
```

```python
model = UANet(input_size=4,hidden_size=20)
```

```python
epochs = 10
optimizer = torch.optim.AdamW(model.parameters(),lr = 0.01)
loss_fn   = torch.nn.MSELoss()
model.train()

for epoch in range(epochs):

	for batch in tqdm(train_loader):
		input , label = batch
		
		optimizer.zero_grad()
		output = model(input.float())
		
		loss = loss_fn(output,label.unsqueeze(1).float())

		loss.backward()
		optimizer.step()

	print(f"Epoch: {epoch + 1}  Loss Per epoch: {loss.item()}")
	
```

```python
model.eval()

with torch.no_grad():

	for batch in tqdm(test_loader):
		input, label = batch
		label = label.unsqueeze(1).float()
		output = model(input)

		loss = loss_fn(output,label)

	print(f"Loss: {loss.item()}")
```
